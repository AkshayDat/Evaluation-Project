# Splitting The data into training and testing sets

```python
x=base_df.iloc[:,1:]
```

```python
x.head()
```

```python
y=base_df.iloc[:,0]
```

```python
y.head()
```

```python
print(x.shape,  y.shape)
```

```python
#Scaling the data in x
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df_x=sc.fit_transform(x)
```

```python
x=df_x
print(x)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split
```

```python
#Finding the best Random state

max_score=0
final_rstate=0

for i in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    lm=LinearRegression() 
    lm.fit(x_train,y_train)
    lm.score(x_train,y_train)
    pred=lm.predict(x_test)
    r_score=r2_score(y_test,pred)
    print("R2 SCORE is equal to :",r_score,"at random state",i)
    
    if r_score > max_score:
        max_score=r_score
        final_rstate=i
        
        
    
    
print("The maximum r2_score is ",max_score, "at random state",final_rstate) 
```

```python
#finalizing the train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=86)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
#applying regularization techniques to overcome underfit and overfit 
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
print("\n")
preed=ls.predict(x_test)
print("Predicted result",preed)
print("\n \n")
print("Actual win \n",y_test)
r_score=r2_score(y_test,preed)
print(r_score)
```

```python
ri=Ridge(alpha=0.0001)
ri.fit(x_train,y_train)
print(ri.score(x_train,y_train))
print("\n")
predd=ri.predict(x_test)
print("Predicted result",predd)
print("\n \n")
print("Actual win \n",y_test)
r_score=r2_score(y_test,predd)
print(r_score)
```

```python
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.0001)
enr=ElasticNet()
enr.fit(x_train,y_train)
enrpred=enr.predict(x_test)
print(enr.score(x_train,y_train))
print("Predicted result",enrpred)
print("\n \n")
print("Actual win \n",y_test)
r_score=r2_score(y_test,enrpred)
print(r_score)
enr.coef_
```
