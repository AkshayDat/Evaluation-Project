# Bivariate Analysis

```python
#plotting barplot to see which student has score the highest marks in each subject
collist=df_std_marks.columns.values
for i in range(3,len(collist)):
    plt.figure(figsize=(10,5))
    plt.show(sns.barplot(df_std_marks['Name'],df_std_marks[collist[i]]))
    i+1
```

```python
#finding marks count of male and female in each subject  
for i in range(3,len(collist)):
    plt.figure(figsize=(10,5))
    plt.show(sns.countplot(df_std_marks[collist[i]],hue=df_std_marks['Gender']))
    i+1
```

```python
#using lineplot to see how the students have performed in each subject
for i in range(3,len(collist)):
    plt.figure(figsize=(10,5))
    plt.show(sns.lineplot(df_std_marks['Name'],df_std_marks[collist[i]]))
    i+1
```
